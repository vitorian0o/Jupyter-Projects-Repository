```python
"""
Exercício Análise de Números

Objetivo: Desenvolver um programa que solicita ao usuário a entrada de 
um número e, com base nesse número, realiza as seguintes operações:

    1. Mostrar o número informado.
    2. Informar se o número é par ou ímpar.
    3. Informar se o número é positivo, negativo ou zero.
    4. Se o número for positivo, calcular e mostrar sua raiz quadrada.

"""

num = float(input("Insira um número: "))

print(num)

if num % 2 == 0:
    print("Par")
else:
    print("Ímpar")
    
if num != 0:
    if num < 0:
        print("Negativo")
    else:
        print("Positivo")
        raiz_quadrada = num**0.5
        print(f"A raiz quadrada de {num} é {raiz_quadrada}")
else:
    print("Zero")
```

```python
"""
Exercício Operações Matemáticas Básicas

Objetivo: Desenvolver um programa que solicite ao usuário a entrada
de dois números e, com base nesses números, realize as seguintes operações:

    1. Calcular e mostrar a soma dos dois números.
    
    2. Calcular e mostrar a subtração do primeiro número pelo segundo.
    
    3. Calcular e mostrar a multiplicação dos dois números.
    
    4. Calcular e mostrar a divisão do primeiro número pelo 
    segundo (se o segundo número não for zero).
    
    5. Verificar e informar se algum dos números (ou ambos) é zero.
    
    6. Calcular e mostrar a média dos dois números.
    
    7. Comparar os dois números e informar qual é maior ou se são iguais.
    
"""

num1 = float(input("Insira o 1º número: "))
num2 = float(input("Insira o 2º número: "))
print(f"Números informados: {num1} e {num2}")

if num1 == 0 and num2 == 0:
    print("Ambos os números são zero")
elif num1 == 0:
    print("O 1º número é zero")
elif num2 == 0:
    print("O 2º número é zero")
else:
    print("Nenhum dos números é zero")
    
def soma(num1, num2):
    result = num1 + num2
    return result

def subtracao(num1, num2):
    result = num1 - num2
    return result

def multiplicacao(num1, num2):
    result = num1 * num2
    return result

def divisao(num1, num2):
    if num2 > 0:
        result = num1 / num2
        return result
    
def media(num1, num2):
    result = soma(num1, num2)
    result = result / 2
    return result

def comparar(num1, num2):
    if num1 != num2:
        if num1 > num2:
            message = f"O 1º número [{num1}] é maior que o 2º número [{num2}]"
        else:
            message = f"O 2º número [{num2}] é maior que o 1º número [{num1}]"
    else:
        message = f"Os números são iguais"
        
    return message

print(f"Soma: {soma(num1, num2)}")
print(f"Subtração: {subtracao(num1, num2)}")
print(f"Multiplicação: {multiplicacao(num1, num2)}")
print(f"Divisão: {divisao(num1, num2)}")
print(f"Média: {media(num1, num2)}")
print(f"Comparação: {comparar(num1, num2)}")
```
